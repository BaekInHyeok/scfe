```python
#weatherAUS.csv를 weather로 읽어서 NA값을 체크해보세요
weather = pd.read_csv("weatherAUS.csv")
weather2 = weather.dropna()
#weather의 NA값을 삭제한 결과를 weather2로 만들어보세요
#weather2의 MinTemp, MaxTemp, Rainfall,	Evaporation,	Sunshine 만 선택해서
#weather3으로 만드세요
weather3 = weather2[ ["MinTemp", "MaxTemp", "Rainfall",	"Evaporation",	"Sunshine"] ]
#weather3를 scaling하고 데이터프레임으로 만드세요
weather3 = pd.DataFrame( scaler.fit_transform( weather3 ), columns = weather3.columns, index=weather3.index )
#weather3, weather2의 RainTomorrow를 8:2로 파티셔닝 하세요
#각각 X_train, X_test, y_train, y_test로 할당하세요
X_train, X_test, y_train, y_test=train_test_split( weather3, weather2.RainTomorrow, test_size=0.2 )
#X_train의 corr()을 적용해서 히트맵으로 그려보세요.
sns.heatmap( X_train.corr() )
#y_train과 y_test의 value_counts를 비교하세요
y_train.value_counts()
y_test.value_counts()
```
